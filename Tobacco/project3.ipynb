```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('datafile.csv')
df.head()
```

isna() checks for missing values (including NaN, None, and others)
isnull() checks for completely empty cells (including missing values and no data).

```python
selected_rows = df.loc[df.iloc[:, 1] == 'Urban']
print(selected_rows['State/UT'])
```

```python
removeindexofindia = df[df['State/UT'] == 'India'].index
df.drop(removeindexofindia, inplace=True)
temp = pd.DataFrame({
    "State/UT": selected_rows['State/UT'],
    "Bidi users in %": selected_rows['Current bidi users (%)']
})
# temp
```

```python
plt.figure(figsize=(20, 10))

sns.barplot(x="State/UT", y = "Bidi users in %", data=temp)
plt.xticks(rotation=90)
plt.show()
```
